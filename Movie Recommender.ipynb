Movie Recommendation Program

```python
import pandas as pd
import numpy as np

rating=pd.read_csv('https://raw.githubusercontent.com/Arnavsmayan/Movie-Recommender/main/ratings.dat',sep='::',engine='python')
rating.columns=['userId','movieId','rating','timestamp']
movie=pd.read_csv('https://raw.githubusercontent.com/Arnavsmayan/Movie-Recommender/main/movies.dat',encoding='latin',sep='::',engine='python')
movie.columns=['movieId','title','genres']
data=pd.merge(rating,movie)
data.drop(['timestamp','movieId'],axis=1,inplace=True)
data.head()

```

```python
avg_reviews=pd.DataFrame(data.groupby('title')['rating'].mean())
avg_reviews['No. of Reviews']=pd.DataFrame(data.groupby('title')['rating'].count())
avg_reviews.rename(columns={'rating':'Avg. Rating'},inplace=True)
avg_reviews.head()
```

```python
user_database=data.pivot_table(index='userId',columns='title',values='rating')
user_database.head()
```

Input The Sample Movie Here:-

```python
relation=user_database.corrwith(user_database["Toy Story (1995)"]) 
relation.dropna(inplace=True)
relation.head()
```

```python
relation=pd.DataFrame(relation,columns=['Percentage User Correlation'])
relation=relation.merge(avg_reviews,on='title',how='left')
relation.drop('Avg. Rating',axis=1,inplace=True)
relation=relation[relation['No. of Reviews']>500].sort_values('Percentage User Correlation',ascending=False)
relation.head()
```

```python
final_set=pd.merge(relation,movie,on='title',how='left')
final_set.drop('movieId',axis=1,inplace=True)
final_set.head(10)
```
